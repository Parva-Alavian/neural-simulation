```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set global Matplotlib settings for Illustrator-friendly figures
plt.rcParams.update({
    "svg.fonttype": "none",    # Keeps text as real text (not outlines)
    "pdf.fonttype": 42,        # Ensures fonts are editable in PDFs
    "font.family": "Arial",    # Use a standard, widely supported font
    "axes.linewidth": 1.5,     # Make axes lines thick enough for visibility
    "xtick.major.size": 5,     # Adjust tick size for clarity
    "ytick.major.size": 5,
})

# Generate sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create figure and axes
fig, ax = plt.subplots(figsize=(6, 4))  # Set figure size (in inches)
ax.plot(x, y, linewidth=2, color="blue", label="Sine Wave")

# Customize axes and labels
ax.set_xlabel("X Axis", fontsize=14)  
ax.set_ylabel("Y Axis", fontsize=14)
ax.set_title("Illustrator-Ready Plot", fontsize=16)
ax.legend(fontsize=12)

# Save as fully editable SVG
fig.savefig("figure_editable.svg", format="svg", metadata={"Title": "Editable Plot"})

# Save as PDF for linking in Illustrator
fig.savefig("figure_editable.pdf", format="eps", metadata={"Title": "Editable Plot"})

plt.show()
```
